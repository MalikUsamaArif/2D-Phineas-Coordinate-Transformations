## Shear by  -3 factor

```python
import matplotlib.pyplot as plt
import numpy as np

# Define shapes
hair = [(-10, 8), (-11, 10), (-13, 9), (-12, 11), (-14, 12), (-12, 13), (-11, 12),
        (-10, 14), (-8, 16), (-9, 13), (-8, 12), (-9, 9)]

face = [(-9, 9), (-8, 4), (-7, 1), (-8, 0.8), (-9, 0), (-8.5, -1), (-7, -1.3), (-5, -5),
        (-4, -8), (-3, -11), (0, -11), (3, -10), (8, -7), (13, -2), (16, 0), (7.5, 2),
        (5, 1.2), (3, 1.8), (0, 1), (-1.4, 2.3), (-2.4, 4), (-3, 6)]

beye = [(-2, 3), (-3, 6), (-1, 7), (1, 7.5), (3, 6), (4.5, 6.8), (5.4, 7.5),
        (6.5, 5), (7.2, 2), (5, 1.2), (3, 1.8), (0, 1)]

eye1 = [(0, 3), (0, 5), (2, 6), (3, 4), (2, 2)]
eye2 = [(3, 3), (4, 5), (5, 5), (6, 3), (5, 2), (4, 2)]
lip = [(5.2, -8.8), (4, -9.1), (3, -9), (2, -8.5), (1, -7), (2, -8.5), (3, -9), (4, -9.1)]

# Transformation: Shear along x-axis by factor -3
def shear_x(coords, factor):
    matrix = np.array([[1, factor], [0, 1]])
    return [tuple(matrix @ np.array([x, y])) for x, y in coords]

# Plotting function
def draw_shape(coords, color='orange', label='Shape', edgecolor='black'):
    shape = np.array(coords + [coords[0]])
    plt.fill(shape[:, 0], shape[:, 1], color=color, alpha=0.8, edgecolor=edgecolor, label=label)
    plt.plot(shape[:, 0], shape[:, 1], 'o-', color=edgecolor)

# Apply shear
hair_s = shear_x(hair, -3)
face_s = shear_x(face, -3)
beye_s = shear_x(beye, -3)
eye1_s = shear_x(eye1, -3)
eye2_s = shear_x(eye2, -3)
lip_s = shear_x(lip, -3)

# Set up plot
plt.figure(figsize=(12, 10))
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(np.arange(-50, 51, 1))
plt.yticks(np.arange(-20, 21, 1))

# Draw sheared shapes
draw_shape(hair_s, color='red', label='Hair', edgecolor='black')
draw_shape(face_s, color='peachpuff', label='Face', edgecolor='black')
draw_shape(eye1_s, color='black', label='Eye 1', edgecolor='black')
draw_shape(eye2_s, color='black', label='Eye 2', edgecolor='black')
draw_shape(beye_s, color='white', label='Eye', edgecolor='black')
draw_shape(lip_s, color='black', label='Lip', edgecolor='black')

plt.title('Phineas After Shear (X-axis, k = -3)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.axis('equal')
plt.xlim(-70, 35)
plt.ylim(-16, 20)
plt.show()
```

# Compression along x-axis by factor of 1/2


```python
# Compression along x-axis by factor of 1/2
def compress_x(coords, factor):
    matrix = np.array([[factor, 0], [0, 1]])
    return [tuple(matrix @ np.array([x, y])) for x, y in coords]

# Apply shear
hair_s = shear_x(hair, -3)
face_s = shear_x(face, -3)
beye_s = shear_x(beye, -3)
eye1_s = shear_x(eye1, -3)
eye2_s = shear_x(eye2, -3)
lip_s = shear_x(lip, -3)

# Then compress along x-axis (x = x * 0.5)
hair_final = compress_x(hair_s, 0.5)
face_final = compress_x(face_s, 0.5)
beye_final = compress_x(beye_s, 0.5)
eye1_final = compress_x(eye1_s, 0.5)
eye2_final = compress_x(eye2_s, 0.5)
lip_final = compress_x(lip_s, 0.5)

# Set up plot
plt.figure(figsize=(12, 10))
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(np.arange(-30, 31, 1))
plt.yticks(np.arange(-20, 21, 1))

# Draw final (sheared + compressed) shapes
draw_shape(hair_final, color='red', label='Hair', edgecolor='black')
draw_shape(face_final, color='peachpuff', label='Face', edgecolor='black')
draw_shape(eye1_final, color='black', label='Eye 1', edgecolor='black')
draw_shape(eye2_final, color='black', label='Eye 2', edgecolor='black')
draw_shape(beye_final, color='white', label='Eye', edgecolor='black')
draw_shape(lip_final, color='black', label='Lip', edgecolor='black')

plt.title('Phineas After Shear (k = -3) + Compression (x * ½)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.axis('equal')
plt.xlim(-40, 30)
plt.ylim(-20, 20)
plt.show()
```
